```python
# Clean the data set from stop words
# your code
from nltk.corpus import stopwords
import nltk


nltk.download('stopwords')
russian_stopwords = stopwords.words('russian')

def remove_stopwords(text):
    if isinstance(text, str):

        words = nltk.word_tokenize(text, language="russian")

        filtered_words = [word for word in words if word.lower() not in russian_stopwords]

        return ' '.join(filtered_words)
    return text

ds['text_clean'] = ds['text'].apply(remove_stopwords)

print(" ")
print(ds['text'].iloc[0][:200] + "...")
print(" ")
print(ds['text_clean'].iloc[0][:200] + "...")
```

```python

comparison_df = pd.DataFrame({
    'original': ds['text'].str.slice(0, 200) + "...",
    'cleaned': ds['text_clean'].str.slice(0, 200) + "...",
    'title': ds['title']
})


display(comparison_df.head(15))
```
